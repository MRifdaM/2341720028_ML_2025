## Soal 2 (5 poin)
1. Lakukan proses data imputation pada fitur yang memiliki data yang hilang
2. Cek kembali apakah masih terdapat data yang hilang

```python
# Jawab Soal 2
# Kerjakan pada cell ini
# Anda diperbolehkan menambah cell jika diperlukan

df['workclass'].fillna(value=df['workclass'].mode, inplace=True)
df['occupation'].fillna(value=df['occupation'].mode, inplace=True)
df['native-country'].fillna(value=df['native-country'].mode, inplace=True)
```

```python
print("Melihat Jumlah Data yang Hilang setelah imputasi:")
df.isnull().sum()
```

## Soal 3 (10 poin)
Inspeksi semua fitur kualitatif. Jika terdapat value yang **tidak sesuai**, **ganti dengan 'Others'** atau yang sesuai atau jika terdapat duplikasi karena **kesalahan penulisan**, lakukan penyesuaian.

```python
# Jawab Soal 3
# Kerjakan pada cell ini
# Anda diperbolehkan menambah cell jika diperlukan

anomaly_cols = df.select_dtypes(include=['object','category']).columns

df[df.eq('?').any(axis=1)].head(20)
```

```python
print("Melihat Jumlah Data yang mengandung value '?' di setiap kolom:")
for i in anomaly_cols:
    print(i, (df[i] == '?').sum())
```

```python
df[anomaly_cols] = df[anomaly_cols].replace('?', 'Others')
print("Melihat Jumlah Data yang mengandung value '?' di setiap kolom:")
for i in anomaly_cols:
    print(i, (df[i] == '?').sum())
```

```python
#Memperbaiki format value pada variabel income
df['income'] = df['income'].replace({'<=50K.': '<=50K', '>50K.':'>50K'})

df.head(300)
```

## Soal 1 - Visualisasi Data (20 poin)
Lakukan inspeksi visual pada,
1. Pada kolom 'age' dengan menggunakan histrogram
2. Pada kolom 'education' education menggunakan barchart
3. Pada kolom 'income' terhadap 'hours_per_week' menggunakan boxplot (kelompokkan berdasarkan kelompok income)

```python
# Jawab 1.1 - Histrogram
num_cols = ['age']
for col in num_cols:
    plt.figure()
    plt.hist(df[col], bins=30)
    plt.title(f'Distribusi {col}')
    plt.xlabel(col); plt.ylabel('Frekuensi')
    plt.show()
```

```python
# Jawab 1.2 - Barchart
plt.figure(figsize=(18,6))
sns.countplot(data=df, x='education')
plt.title('Education')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.show()
```

```python
# Jawab 1.3 - Boxplot

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='income', y='hours-per-week')
plt.title('Hours-per-Week berdasarkan Income')
plt.xlabel('Income')
plt.ylabel('Hours per Week')
plt.show()
```

```python
# Jawab 1.4 - Lineplot
plt.figure()
sns.lineplot(x='age', y='capital-gain', data=df, label='Capital Gain')
sns.lineplot(x='age', y='capital-loss', data=df, label='Capital Loss')
plt.title('age terhadap capital-gain dan capital-loss')
plt.xlabel('Age')
plt.ylabel('Capital')
plt.legend()
plt.show()
```
