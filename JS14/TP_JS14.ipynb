```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, optimizers
import matplotlib.pyplot as plt

# 1. Load & Preprocess Data
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

# 2. Definisi Model yang Lebih Kompleks (VGG-Style Block)
model = models.Sequential()

# Block 1
model.add(layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(32, (3, 3), padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.2)) # Mencegah Overfitting

# Block 2
model.add(layers.Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(64, (3, 3), padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.3))

# Block 3
model.add(layers.Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Conv2D(128, (3, 3), padding='same', activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.4))

# Flatten & Dense Layers
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax')) # Gunakan softmax untuk output probability

# 3. Compile Model
# Menggunakan optimizer Adam dengan learning rate spesifik kadang membantu
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 4. Training dengan Data Augmentation
# Kita gunakan ImageDataGenerator untuk augmentasi secara real-time
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
)

# Hitung langkah per epoch
it_train = datagen.flow(train_images, train_labels, batch_size=64)
steps = int(train_images.shape[0] / 64)

print("\n--- Mulai Training (Bisa memakan waktu 15-30 menit) ---")
# Epochs ditingkatkan karena model lebih kompleks butuh waktu belajar lebih lama
history = model.fit(it_train, steps_per_epoch=steps, epochs=30,
                    validation_data=(test_images, test_labels))

# 5. Evaluasi
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f"\nAkurasi Akhir: {test_acc:.2%}")

# Plotting
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Akurasi Model Modifikasi')
plt.show()
```

### 2. Metodologi yang saya coba lakukan?

#### A. Arsitektur yang Lebih Dalam (VGG-Style)
Lihat pola pada layer: `32 -> 64 -> 128`.
Kita menumpuk filter konvolusi dengan jumlah yang semakin besar.
* **Layer awal (32 filter):** Belajar fitur sederhana (garis, tepi, warna).
* **Layer tengah (64 filter):** Menggabungkan garis menjadi bentuk (mata, telinga, roda).
* **Layer akhir (128 filter):** Mengenali objek kompleks (wajah kucing vs wajah anjing).
Tanpa kedalaman ini, model tidak cukup "cerdas" untuk membedakan 10 kelas CIFAR yang rumit.

#### B. Batch Normalization (Kunci Kestabilan)
Di log, setelah setiap `Conv2D`, ada `BatchNormalization`.
* **Fungsinya:** Menormalkan output dari layer sebelumnya agar sebaran datanya stabil.
* **Efeknya:** Model belajar jauh lebih cepat dan lebih stabil. Tanpa ini, mencapai 80% akan butuh waktu training 2x atau 3x lebih lama, atau bahkan gagal konvergen.

#### C. Dropout (Anti-Hafalan) memiliki layer `Dropout` setelah setiap blok pooling dan di layer dense.
* **Cara kerja:** Mematikan neuron secara acak (misal 20-40% dimatikan sementara saat training).
* **Hasilnya:** Model dipaksa untuk tidak "menghafal" (overfitting) satu ciri saja. Model dipaksa mencari fitur lain. Inilah alasan kenapa akurasi Validasi bisa setinggi akurasi Training.

#### D. Data Augmentation (Tersirat)
Meskipun tidak terlihat di log model, tapi terlihat dari perilaku training (Validation Acc > Training Acc di beberapa titik), kemungkinan besar `ImageDataGenerator` bekerja dengan baik.
* Model tidak hanya melihat gambar kucing yang duduk tegak.
* Model melihat kucing yang sedikit miring, digeser ke kiri, atau dibalik (flip).
* Ini membuat dataset seolah-olah menjadi jauh lebih banyak dan bervariasi.

### Kesimpulan

Mendapatkan akurasi **81.36%** karena berhasil membangun model yang **seimbang**. Modelnya cukup kompleks (Deep) untuk mempelajari fitur sulit, tapi dijaga ketat oleh **Dropout** dan **Batch Normalization** agar tidak "curang" (overfitting).


```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# 1. Load Data MNIST
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()

# 2. Preprocessing
# Reshape data agar memiliki channel (28, 28, 1)
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

# Normalisasi piksel (0-1)
train_images, test_images = train_images / 255.0, test_images / 255.0

# 3. Buat Model CNN
model_mnist = models.Sequential()

# Layer Konvolusi & Pooling
model_mnist.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model_mnist.add(layers.MaxPooling2D((2, 2)))
model_mnist.add(layers.Conv2D(64, (3, 3), activation='relu'))
model_mnist.add(layers.MaxPooling2D((2, 2)))
model_mnist.add(layers.Conv2D(64, (3, 3), activation='relu'))

# Fully Connected Layer
model_mnist.add(layers.Flatten())
model_mnist.add(layers.Dense(64, activation='relu'))
model_mnist.add(layers.Dense(10, activation='softmax'))

# 4. Compile
model_mnist.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

model_mnist.summary()

# 5. Training
print("\n--- Training Model MNIST ---")
history_mnist = model_mnist.fit(train_images, train_labels, epochs=5,
                                validation_data=(test_images, test_labels))

# 6. Evaluasi
test_loss, test_acc = model_mnist.evaluate(test_images, test_labels, verbose=2)
print(f"\nAkurasi MNIST: {test_acc:.2%}")
```

### 1. Analisis Arsitektur Model (Mengapa Model Ini Efisien?)

    * Input asli adalah 28x28 piksel.
    * Setiap melewati `Conv2D` dan `MaxPooling2D`, dimensi spasial (lebar x tinggi) makin kecil, tapi kedalaman (jumlah filter) makin tebal.
    * Perubahan dimensi: `26x26` $\rightarrow$ `13x13` $\rightarrow$ `11x11` $\rightarrow$ `5x5` $\rightarrow$ `3x3`.
    * **Analisis:** Model fokus mengekstrak fitur sederhana (garis lurus/lengkung) di awal, lalu menggabungkannya menjadi fitur kompleks (bentuk angka 8, 0, 6) di layer akhir (3x3).
* **Total Parameter (93,322):**
    * Jumlah parameter sangat kecil (hanya 93 ribu) dibandingkan model CIFAR (bisa jutaan).
    * **Dampaknya:** Model sangat ringan, cepat dilatih (hanya ~1 menit per epoch), dan risiko overfitting lebih kecil karena kapasitas model pas untuk kerumitan data.

### 3. Perbandingan: Mengapa MNIST (99%) Jauh Lebih Mudah dari CIFAR (81%)?
 mungkin bertanya, kenapa di Tugas 1 susah payah pakai Dropout/Batch Normalization cuma dapat 81%, tapi di sini pakai model sederhana langsung 99%?
